```python
#pdf can be computed using counts and frequencies of data points in each window
#very small window is overfitting and larger window is underfitting that how window width affects the pair plot
#it also called density plot and probability plot 
#pdf is called a density plot because it represents how densely data points are packed around a value
#pdf is called probablity distribution because it shows how probablity is distributed
#higher peak is more likely region
#the y axis value is  probability density not probability
#disadvantage of pdf can we say what percentage of versicolor points have petal length of less than 5
#guassian distribution and normal distribution is same .it is called normal because it is most frequent distribution
#in cdf we can visualize what percentage of flower is less than a specific value
#np.histogram gives two things it gives number of data points in each bin and bin_edges gives boundaries of those bins
counts,bin_edges=np.histogram(data['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()



```

```python
counts,bin_edges=np.histogram(data['age'],bins=20,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
print(max(pdf))
```
